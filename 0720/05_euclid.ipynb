유클리드 호제법을 이용한 최대공약수와 최소공배수 계산하기  
숫자 2개를 입력받아 큰수와 작은수를 구분한다.  
큰수를 작은수로 나눈 나머지가 0이면 작은수가 최대공약수가 되고 큰수를 작은수로 나눈 나머지가 0이 아니면 큰수를 기억하던 기억장소에 작은수를 넣고 작은수를 기억하던 기억장소에 나머지를 넣어준 후 작업을 반복한다.  
최대공약수가 계산되면 입력받은 숫자 2개를 곱한 결과를 최대공약수로 나눠주면 최소공배수가 계산된다.

```python
from IPython.display import Image
Image('./euclid.png', width='800')
```

```python
# 숫자 2개를 입력받아 큰수와 작은수를 판단한다.
a, b = map(int, input('숫자 2개를 입력하세요: ').split())
if a > b:
    big = a; small = b
else:
    big = b; small = a
# print('big = {}, small = {}'.format(big, small))

# 입력받은 숫자에 따라 반복 횟수가 달라지므로 무한루프를 돌려 처리한다.
while True:
    r = big % small # 큰수를 작은수로 나눈 나머지를 계산한다.
    if r == 0:      # 무한루프 종료 조건
        break
    # ===== if r == 0
    # 다음 연산을 위해 작은수를 큰수로 나머지를 작은수롤 옮겨준다.
    big = small
    small = r
# ===== while True

# 최소공배수를 계산하고 최대공약수와 최소공배수를 출력한다.
print('최대공약수: {}, 최소공배수: {}'.format(small, int(a * b / small)))
```

```python
a, b = map(int, input('숫자 2개를 입력하세요: ').split())
# r은 큰수를 작은수로 나눈 나머지를 기억하는 변수
r = 1
if a > b:
    high = a; low = b
else:
    low = a; high = b
    
# 최초 실행시 변수 r에 저장된 값이 0보다 크므로 나머지를 계산하는 문장이 실행된다.
while r > 0:
    r = high % low
    high = low
    low = r
# ===== while r > 0

l = int(a * b / high)
print('최대공약수: {}, 최소공배수: {}'.format(high, l))
```
