```python
lm = 0 #     =>     1 .
#   4 ~ 10000       (n) 4 ~ 10000  1 .
for n in range(4, 10001):
    
    #                0 .
    total = 0 #       
    #               .
    k = n // 2
    
    #   n        .
    for j in range(1, k + 1):
        #       n 1 ~ k     .
        r = n % j
        # n j   0 j n    .
        if r == 0:     # j n ?
            total += j # j n    .
        # ===== if r == 0
    # =====for j in range(1, k + 1)
    
    # n    n   (lm) 1.
    if n == total: # ?
        lm += 1    #   1.
        print('{0}  => {1:4d}'.format(lm, n))
    # if n == total
    
# ===== for n in range(4, 10001)
print('  {} .'.format(lm))
```

```python
lm = 0
for n in range(4, 10001):
    total = 0
    k = n // 2
    for j in range(1, k + 1):
        r = n % j
        if r == 0:
            total += j
    if n == total:
        lm += 1
        print('{0}  => {1:4d}'.format(lm, n))
print('  {} .'.format(lm))
```
