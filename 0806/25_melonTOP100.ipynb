```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
request = requests.get(targetSite)
print(request) # <Response [406]> => 헤더 정보 문제로 사이트의 요청 정보를 얻어올 수 없다.
# 406(허용되지 않음): 요청한 페이지가 요청한 콘텐츠 특성으로 응답할 수 없다.
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
# requests 모듈의 get() 메소드로 사이트의 정보를 요청했을때 406 에러가 발생되면 아래와 같이 headers 속성을 지정한다.
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers=header)
# print(request) # <Response [200]>
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
# 노래 제목은 'class' 속성이 'ellipsis rank01'인 <div> 태그 내부에 있다.
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
# print(titles)
# print(len(titles))
for title in titles:
    # print(title.text.strip())
    # print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
# 가수 이름은 'class' 속성이 'checkEllipsis'인 <span> 태그 내부에 있다.
artists = soup.findAll('span', {'class': 'checkEllipsis'})
# print(artists)
# print(len(artists))
for artist in artists:
    # print(artist.text.strip())
    # print(artist.text.split('\n')[0])
    print(artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    print('{0:3d}등 {1} - {2}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
# file = open('./data/melonTOP100.txt', 'w')
file = open('./data/melonTOP100.txt', 'w', -1, 'UTF-8')
file.write('{} 현재 Melon 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    data = '{0:3d}등 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')
file.close()
print('melonTOP100.txt 파일에 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 출력한다.
try:
    # file = open('./data/melonTOP100.txt', 'r')
    file = open('./data/melonTOP100.txt', 'r', -1, 'UTF-8')
    # readlines() 메소드는 텍스트 파일의 내용 전체를 한꺼번에 읽어서 리스트로 얻어온다.
    lines = file.readlines()
    for line in lines:
        print(line.strip())
except FileNotFoundError:
    print('melonTOP100.txt 파일이 디스크에 없습니다.')
```
