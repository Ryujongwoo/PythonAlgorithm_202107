```python
class Stack:
    # Stack 클래스가 생성될 때 생성자로 스택의 크기가 전달되면 전달된 크기만큼의 기억 공간을 가지는 스택을 생성하고 크기를
    # 전달받지 못하면 기본값(default)으로 지정한 5개의 데이터를 저장할 수 있는 스택을 만든다. => 디폴트 인수를 사용한다.
    def __init__(self, size = 5): # 디폴트 인수는 '변수이름 = 기본갑' 형태로 사용한다.
        # print('Stack 클래스가 생성될 때 자동으로 실행됩니다.')
        # print('Stack 클래스 객체가 생성된 주소: {}'.format(self))
        # print('스택의 크기: {}'.format(size))
        # 생성자 함수에서 스택을 만든다.
        # 변수 이름 앞에 'self.'가 붙은 변수를 멤버 변수라 부르며 클래스 내부의 모든 메소드에서 사용할 수 있고 'self.'이 
        # 붙지않은 변수를 지역 변수라 부르며 변수가 선언된 함수 내부에서만 사용할 수 있다.
        self.stack = []  # 스택으로 사용할 빈 리스트 => 빈 리스트에 데이터는 append() 메소드로 추가한다.
        self.size = size # 스택의 크기
        # 스택에 데이터가 입력되면 top은 1증가하고 데이터가 출력되면 1감소한다.
        self.top = 0     # top, SP(Stack Pointer) => 스택에 몇 개의 데이터가 저장되었나 확인한다.
        
    # push => 입력
    def push(self, data):
        # 스택에 중복되는 데이터의 입력을 허용하지 않는다.
        if data in self.stack: # 입력하려는 데이터가 스택에 존재하는가?
            # 추가하려는 데이터가 스택에 존재하기 때문에 중복되는 데이터라고 메시지를 출력한다.
            print('{}는(은) 중복되는 데이터입니다.'.format(data))
        else:
            # 스택에 추가하려는 데이터가 스택에 존재하지 않으므로 스택에 데이터를 추가한다.
            # 데이터를 추가하기 전에 overflow 인가 검사한다.
            # 스택의 크기(self.size)가 5일때 스택으로 사용할 리스트의 인덱스(self.top)는 0, 1, 2, 3, 4만 사용할 수 있다.
            if self.size > self.top:
                # overflow가 발생되지 않았으므로 스택에 데이터를 저장한다.
                self.stack.append(data)
                # 스택에 데이터를 저장했으므로 top을 1증가시킨다.
                self.top += 1
                print('스택에 {}을(를) 저장합니다.'.format(data))
            else:
                # overflow가 발생되면 스택이 가득찼다는 메시지를 출력한다.
                print('overflow 발생... 스택이 가득차서 {}를(을) 저장할 수 없습니다.'.format(data))
            # ===== if self.size > self.top
        # ===== if data in self.stack
        # 스택에 저장된 데이터를 출력하는 함수를 실행한다.
        # 현재 클래스의 다른 함수를 실행할 때는 반드시 앞에 self를 붙여서 실행한다.
        self.view()
    
    # pop => 출력
    def pop(self):
        # underflow 인가 검사한다.
        if self.top <= 0:
            print('스택에 저장된 데이터가 없습니다.')
        else:
            # 파이썬의 리스트 관련 메소드 중에서 pop() 메소드를 사용해서 스택에 저장된 데이터를 얻어온 후 리스트에서 제거한다.
            data = self.stack.pop()
            # 스택에 저장된 데이터가 출력되었으므로 top을 1감소시킨다.
            self.top -= 1
            print('pop 데이터: {}'.format(data), end=', ')
            self.view()
    
    # view => 보기
    def view(self):
        # 스택에 저장된 데이터를 출력한다.
        print('스택에 저장된 데이터 => ', end='')
        # underflow 인가 검사한다.
        if self.top <= 0:
            # 스택에 저장된 데이터가 없으므로 없다고 출력한다.
            print('없음', end='')
        else:
            # 스택에 저장된 데이터의 개수만큼 반복하며 스택에 저장된 데이터를 출력한다.
            for i in range(self.top):
                print(self.stack[i], end=' ')
        # ===== if self.top <= 0
        print()
```

```python
stack = Stack()
stack.view()
stack.push(111)
stack.push(111) # 중복 데이터
stack.push(333)
stack.push(222)
stack.push(777)
stack.push(555)
stack.push(999) # overflow
print('=' * 80)
stack.view()
print('=' * 80)
stack.pop()
stack.pop()
stack.pop()
stack.pop()
stack.pop()
stack.pop()
```
