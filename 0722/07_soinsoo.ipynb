```python
#    2   .
while True:
    n = int(input('   : '))
    if n < 2:
        print('2     .')
        break
    # ===== if n < 2
    
    #     
    s = [] #      =>   append()   .
    c = 0  #  ,     1 .
    number = n #     .
    
    #   . =>          .
    while True:
        k = 2 #     => 1   2 .
        while True:
            r = n % k
            if r == 0: #   ?
                break  #     .
            # ===== if r == 0
            k += 1     #    k 1      .
        # ===== while True
        
        c += 1      #   1. =>   1     . => 
        s.append(k) #   .
        n //= k     # #        () n .
        if n == 1:  #   ?
            break   #     .
    # ===== while True
    
    #        1() .
    if c == 1:
        print(': {} => '.format(number))
    else:
        print(': {} => '.format(number), end='')
        for i in range(len(s) - 1):
            print(s[i], end=' * ')
        print(s[-1])
```
