```python
# 소인수 분해하려는 숫자가 2미만이면 소인수 분해를 종료한다.
while True:
    n = int(input('소인수 분해할 숫자 입력: '))
    if n < 2:
        print('2보다 작은 숫자가 입력되어 프로그램을 종료합니다.')
        break
    # ===== if n < 2
    
    # 소인수 분해에 사용할 기억장소 초기화
    s = [] # 분해한 소인수를 기억할 빈 리스트 => 데이터 추가는 append() 메소드를 사용해야 한다.
    c = 0  # 소인수의 개수, 소인수가 발생될 때 마다 1씩 증가시킨다.
    number = n # 입력받은 소인수 분해할 숫자를 저장해둔다.
    
    # 소인수 분해를 시작한다. => 입력된 숫자의 크기에 따라 반복 횟수가 다르므로 무한루프를 돌려 처리한다.
    while True:
        k = 2 # 소인수 분해를 시작할 수 => 1은 소수가 아니므로 2부터 시작한다.
        while True:
            r = n % k
            if r == 0: # 소인수 분해가 되었는가?
                break  # 소인수 분해가 되었으먼 무한루프를 탈출한다.
            # ===== if r == 0
            k += 1     # 소인수 분해가 안되었으므로 k를 1증가 시킨 후 다시 소인수 분해를 실행한다.
        # ===== while True
        
        c += 1      # 소인수의 개수를 1증가시킨다. => 소인수의 개수가 1개면 자기 자신으로 나눠서 분해된 것이다. => 소수
        s.append(k) # 소인수를 리스트에 추가한다.
        n //= k     # # 다음 소인수를 계산하기 위해서 소인수 분해하고 남은 숫자(몫)를 n에 넣어준다.
        if n == 1:  # 소인수 분해가 끝났는가?
            break   # 소인수 분해가 완료되면 무한루프를 탈출한다.
    # ===== while True
    
    # 소인수 분해 결과를 출력하기 위해서 소인수의 개수가 1인가(소수인가) 비교한다.
    if c == 1:
        print('입력: {} => 소수'.format(number))
    else:
        print('입력: {} => '.format(number), end='')
        for i in range(len(s) - 1):
            print(s[i], end=' * ')
        print(s[-1])
```
